```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
import argparse
import PIL.ImageDraw as ImageDraw
import PIL.Image as Image
# Python 2/3 compatibility
from __future__ import print_function

import cv2 
import video
from skimage.measure import compare_ssim
import imutils

#import subproces


%matplotlib inline


json_file = open("/home/ankit/Downloads/Cam_02_North_Ave_EB_Spring_St.json", "r", encoding="utf-8")
sh_json= json.load(json_file)
json_file.close()

dots=[]
image = Image.new("RGB", (640, 480))

draw = ImageDraw.Draw(image)
#print(type( sh_json['Costco_Front_Image_0013.jpg193064']))
for person in sh_json:
    y=len(sh_json['valid_vertices'])
for x in range(0,y):
    k=(sh_json['valid_vertices'])
    dots.append((k[x][0],k[x][1]))
    #print(points)
#draw.polygon((points), fill='White')
#image.show()
#image.save('image1.png')


'''
example to show optical flow

USAGE: opt_flow.py [<video_source>]

Keys:
 1 - toggle HSV flow visualization
 2 - toggle glitch

Keys:
    ESC    - exit
'''



def draw_flow(points,flow,dots, step=16):                                                # steps means in fig  the dotted points how much away from each other check step =100,10,50

    h,  w = points.shape[:2]                                                           # heights  and weights in image (if it is 3 then it also give no. of channel of img)
    y,  x= np.mgrid[step/2:h:step, step/2:w:step].reshape(2,-1).astype(int)           # mesh-grid gives ndarrays all of the same dimensions
    fx, fy = flow[y,x].T
    points1 = np.vstack([x, y,x+fx, y+fy]).T.reshape(-1, 2, 2)                      # vstack is use for to collect the array and make stack of array in vertical form (row wise)
    points1 = np.int32(points1+0.5)                                                # when we add 0.5 its shifts the dotted points/circle 
    vis = cv.cvtColor(points, cv.COLOR_GRAY2BGR)
    # pts = np.array([points1], np.int32)
    temp_pts = np.array(dots, np.int32)
    temp_pts = temp_pts.reshape((-2, 1, 2)) .astype(int)
                                                          #we use func here because we have to visualize the  flow on dotted points 
    cv2.polylines(vis, [temp_pts], True, (0, 255,200), thickness=2)  
    cv.polylines(vis,points1,True,(0,255,200))
    for (x1, y1),(_x2,_y2) in points1:  
        cv.circle(vis, (x1, y1),1,(0, 255,200),-1)                      #(vis#means on image #(x1,y1) tells where region of cicle/points are changes  #1 means diameter of circle # color of circle )     
    return vis
    


def draw_hsv(flow):
    h, w = flow.shape[:2]
    fx, fy = flow[:,:,0], flow[:,:,1]
    ang = np.arctan2(fy, fx) + np.pi
    v = np.sqrt(fx*fx+fy*fy)
    hsv = np.zeros((h, w, 3), np.uint8)
    hsv[...,0] = ang*(180/np.pi/2)
    hsv[...,1] = 255
    hsv[...,2] = np.minimum(v*4, 255)
    bgr = cv.cvtColor(hsv, cv.COLOR_HSV2BGR)
    return bgr


def warp_flow(img, flow):
    h, w = flow.shape[:2]
    flow = -flow
    flow[:,:,0] += np.arange(w)
    flow[:,:,1] += np.arange(h)[:,np.newaxis]
    res = cv.remap(img, flow, None, cv.INTER_LINEAR)
    return res


if __name__ == '__main__':
    import sys    #The sys module provides information about constants, functions and methods of the Python interpreter. 
                   # dir(system) gives of the available constants, functions and methods.
    print(__doc__)

    cam = cv2.VideoCapture('/home/ankit/Downloads/Optical-Flow_Python-master/data/vtest.avi')
    cam.set(cv2.CAP_PROP_FRAME_WIDTH, 1920)
    cam.set(cv2.CAP_PROP_FRAME_HEIGHT, 1080)
    codec = cv2.VideoWriter_fourcc(*'XVID') 
    output1 = cv2.VideoWriter('/home/ankit/Downloads/just.avi', codec, 60.0,(1920,1080))
    ret, prev = cam.read()                                   #cam returns boolean value if it reads frame by frame       # ret, prev are both returned  # ret is the return status   # prev is the image (stored in previous image,     #because we'll collect the "current" image in the loop
    prevgray =cv2.cvtColor(prev, cv.COLOR_BGR2GRAY)         # Convert the scaled image to grey scale # Some gui options (hsv -> hue, saturation value) 
    show_hsv = False                                       # show_glitch -> shows the mapped image, useful to detect glitches #(applies the flow field to the initial image)
    show_glitch = False
    cur_glitch = prev.copy()                               # use this image for glitch detection. 
   
   

    while(cam.isOpened()) :
        ret, prev = cam.read() 
        gray = cv2.cvtColor(prev, cv.COLOR_BGR2GRAY)
        
        flow = cv.calcOpticalFlowFarneback(prevgray, gray, None,   0.5, 3, 15, 3, 5, 1.2, 0)
        cv.imshow('flow',  draw_flow(gray,flow,dots))    #its shows the image
        if show_hsv:
            cv.imshow('flow HSV', draw_hsv(flow))           # draw the hue saturation value 
            
        if show_glitch:                                      # show the glitch image
            cur_glitch = warp_flow(cur_glitch, flow)
            cv.imshow('glitch', cur_glitch)
            
        ch = cv.waitKey(20)                                    #cv::waitKey(n) func is used to introduce a delay of n milliseconds while rendering images to windows.
                                                             #When used as cv::waitKey(0) it returns the key pressed by the user on the active window. 
                                                                #This is typically used for keyboard input from user in   
        if ch == 27:
            break
        if ch == ord('1'):
            show_hsv = not show_hsv
            print('HSV flow visualization is', ['off', 'on'][show_hsv])
        if ch == ord('2'):
            show_glitch = not show_glitch
            if show_glitch:
                cur_glitch = img.copy()
            print('glitch is', ['off', 'on'][show_glitch])
    cam.release()
    output1.release()
    cv.destroyAllWindows()
```
